```python
import numpy as np

# 90度単位で回転する回転行列を列挙する関数
def generate_rotation_matrices():
    matrices = []
    # X軸を中心に90度単位で回転する行列を列挙する
    for i in range(4):
        theta = i * np.pi / 2
        matrix = np.array([
            [1, 0, 0],
            [0, np.round(np.cos(theta)), -np.round(np.sin(theta))],
            [0, np.round(np.sin(theta)), np.round(np.cos(theta))],
        ], dtype=int)
        matrices.append(matrix)

    # Y軸を中心に90度単位で回転する行列を列挙する
    for i in range(4):
        theta = i * np.pi / 2
        matrix = np.array([
            [np.round(np.cos(theta)), 0, np.round(np.sin(theta))],
            [0, 1, 0],
            [-np.round(np.sin(theta)), 0, np.round(np.cos(theta))],
        ], dtype=int)
        matrices.append(matrix)

    # Z軸を中心に90度単位で回転する行列を列挙する
    for i in range(4):
        theta = i * np.pi / 2
        matrix = np.array([
            [np.round(np.cos(theta)), -np.round(np.sin(theta)), 0],
            [np.round(np.sin(theta)), np.round(np.cos(theta)), 0],
            [0, 0, 1],
        ], dtype=int)
        matrices.append(matrix)

    return matrices

# 使用例
mem = set()
matrices = []
for m1 in generate_rotation_matrices():
    for m2 in generate_rotation_matrices():
        for m3 in generate_rotation_matrices():
            m = m1.dot(m2).dot(m3)
            if str(m) in mem:
                continue
            mem.add(str(m))
            matrices.append(m)
```

```python
def get_mat(i: int, j: int, matrices: list) -> list:
    can = []
    for m in matrices:
        if (m.dot(axis[i]) == axis[j]).all():
            can.append(m)
    return can

def get_indices(m) -> list:
    r = []
    for a in axis:
        c = m.dot(a)
        for i, b in enumerate(axis):
            if (c == b).all():
                r.append(i)
                break
    return r
```

```python
xp = np.array([1, 0, 0])
xm = np.array([-1, 0, 0])
yp = np.array([0, 1, 0])
ym = np.array([0, -1, 0])
zp = np.array([0, 0, 1])
zm = np.array([0, 0, -1])
axis = [xp, xm, yp, ym, zp, zm]
```

```python
for i in range(6):
    for j in range(i + 1, 6):
        aa = [get_indices(m) for m in get_mat(i, j, matrices)]
        bb = [get_indices(m) for m in get_mat(j, i, matrices)]
        for a in aa:
            if a in bb:
                print(f'({i}, {j}) | ({j}, {i}) => {a},')
                break
```

```python
from collections import defaultdict
d = defaultdict(list)
for i1 in range(6):
    for j1 in range(6):
        for i2 in range(6):
            for j2 in range(6):
                if (i1, j1) == (i2, j2):
                    continue
                m = get_mat(i2, j2, get_mat(i1, j1, matrices))
                if m:
                    a = str(get_indices(m[0]))
                    d[a].append(str((i1, j1, i2, j2)))
for key, value in d.items():
    a = ' | '.join(value)
    print(f'{a} => {key},')
```
